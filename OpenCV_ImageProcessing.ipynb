## Geometric Transformations

```python
#Translation - shifting of an object's location
#for matrice multiplication, number of columns of first matrix has to be = no of rows of second matrix

#convert to Gray scale
rows, cols = gray.shape
#set the transformation matrix M
M = np.float64([[1,0,100],[0,1,50]])

result = cv2.warpAffine(gray, M,(cols,rows))
cv2_imshow(result)
```

```python
#using CV.INTER_AREA

result = cv2.resize(image, None, fx=0.5, fy =0.5, interpolation=cv2.INTER_AREA)
cv2_imshow(result)
```

```python
#using cv.INTER_LINEAR
result = cv2.resize(image, None, fx=2, fy=0.5, interpolation=cv2.INTER_LINEAR)
cv2_imshow(result)
```

```python
rows,cols
```

```python
#create transformation matrix
#rotate from mid position of image
#rotate by 90 deg
M = cv2.getRotationMatrix2D(((cols-1)/2, (rows-1)/2), 185, 0.5)
result = cv2.warpAffine(image, M, (cols,rows))

cv2_imshow(result)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
```

```python
#(50,250) --> (100,275)
#(200,250) --> (250,300)
#(50,150) --> (100,175)

rows, cols, ch = image.shape
pts1 = np.float32([[50,250],[200,250],[50,150]])
pts2 = np.float32([[100,275],[250,300],[100,175]])

#create transformation matrix M
M = cv2.getAffineTransform(pts1, pts2)
result = cv2.warpAffine(image, M, (cols,rows))
cv2_imshow(result)
```

```python
#Perspective Transformation

# (0,250) -->(100,300)
# (300,250) -->(350,300)
# (0,50) --> (50,100)
# (300,50) --> (350,75)

pts1 = np.float32([[0,250],[300,250],[0,50],[300,50]])
pts2 = np.float32([[100,300],[350,300],[50,100],[350,75]])

#create transformation matrix
M = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(image, M, (cols,rows))
cv2_imshow(result)
```
